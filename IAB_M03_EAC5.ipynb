## 2 (3p) Extreure l'import de la imatge d'un tiquet

Abans de començar la interfície d'usuari programarem i provarem el funcionament del procés que s'executarà per darrera de l'aplicació. Bàsicament aquest procés consisteix en:
1. Pujar i analitzar la imatge.
2. Crear una funció per buscar l'import del camp 'Total' en la resposta.
3. Formatejar el valor trobat per poder-hi treballar.

````python
# Carrega i mostra la imatge d'un tiquet o factura
from IPython.display import Image,display
documentName='imagen1.jpg'
display(Image(filename=documentName))
````

````python
# Converteix la imatge a un array de bytes
with open(documentName,'rb') as document:
  imageBytes=bytearray(document.read())
````

````python
# Analitza la imatge amb l'API de textract extraient els formularis
response = textract.analyze_document(Document={'Bytes':imageBytes},FeatureTypes=['FORMS'])
doc=Document(response)
print(doc.pages[0].form)
````

````python
from botocore.session import paginate
# Crea una funció que retorni el valor del total del tiquet.
# Tingues en compte que el nom de la clau no sempre serà "TOTAL".
# Comprova que funciona amb imatges d'altres tiquets i ves-la millorant.

import re

etiquetas_total=['TOTAL','Total','Total:','Precios IVA Incluido Total']

def total():
  for campo in doc.pages[0].form.fields:
    if campo.key.text in etiquetas_total:
      precio=campo.value.text
      precio=precio.replace(',','.')
      precio = re.sub(r"[^0-9.\s]", ' ', precio)
      total=float(precio)
  return total


print('total=',total())


````

````python
# Defineix i executa una funció per formatar la sortida a un número amb dos decimals
def redondear(num):
  return f"{num:.2f}"

print('total=',redondear(total()))
````

## 3 (3p) Integració a Gradio


És el moment d'integrar el procés a Gradio. Primer crearem l'interfície d'usuari amb *Blocks* per tal que sigui semblant a l'aplicació de l'enunciat, el següent pas serà integrar i modificar, si cal, el codi que hem desenvolupat anteriorment en una funció que s'executarà cada cop que l'usuari pengi una nova imatge.

Cal tenir en compte que el component [*Image*](https://www.gradio.app/docs/gradio/image) pot passar la imatge com a array de NumPy, objecte PIL o una ruta al fitxer temporal, en aquest cas ens serà molt útil aquest darrer cas, i per tant:
```
gr.Image(type='filepath')
```
Per calcular el valor acumulat podem treballar amb [estats de sessió](https://www.gradio.app/guides/state-in-blocks) o amb una variable global que s'actualitza cada cop que detectem un nou import.

Afegeix una captura de pantalla de la teva aplicació al final de l'apartat.

````python
# Crea l'estructura visual de components Gradio amb Blocks
import gradio as gr

gr.close_all() # Cierra todos los puertos abiertos

with gr.Blocks() as tickets:
    gr.Markdown("<h1 style='text-align:center;'>Procesador de tickets</h1>")
    gr.Markdown("<p style='text-align:center;'>Sube un ticket y obtendras el total automáticamente</p>")

    with gr.Row():
      entrada = gr.Number(label="Importe del ticket (€):")
      salida = gr.Number(label="Total acumulado (€):")

    acumulado=gr.State(0.0)
    imagen = gr.Image(type='filepath')
    imagen.upload(fn=predict, inputs=[imagen,acumulado], outputs=[entrada,salida,acumulado])
````

````python
# Afageix les funcions que hem creat anteriorment en una sola funció que s'executa cada cop que es puja una imatge.
# També hauràs de gestionar com s'actualitza el camp de l'import acumulat.

import re

etiquetas_total=['TOTAL','Total','Total:','Precios IVA Incluido Total']

def predict(archivo,acumulado):

    # Leemos el fichero y lo convertimos en un array de bytes
    with open(archivo,'rb') as document:
      imageBytes=bytearray(document.read())

    # Analizamos los documentos
    response = textract.analyze_document(Document={'Bytes':imageBytes},FeatureTypes=['FORMS'])
    doc=Document(response)

    # Buscamos coincidencias y sumamos el resultado al acumulado
    for campo in doc.pages[0].form.fields:
      if campo.key.text in etiquetas_total:
        precio=campo.value.text
        precio=precio.replace(',','.')
        precio = re.sub(r"[^0-9.\s]", ' ', precio)
        total=float(precio)
        acumulado+=total
        return f'{total:.2f}',f'{acumulado:.2f}',acumulado

tickets.launch()
````
